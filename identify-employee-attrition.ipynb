# **✅ Mapping the string values into integer**
***

```python
d = {'Non-Travel': 1, 'Travel_Frequently': 2, 'Travel_Rarely': 3}
attrition['BusinessTravel'] = attrition["BusinessTravel"].map(d)

d = {'Human Resources': 1, 'Research & Development': 2, 'Sales': 3}
attrition['Department'] = attrition['Department'].map(d)

d = {'Human Resources': 1, 'Life Sciences': 2, 'Marketing': 3, 'Medical': 4, 'Other': 5, 'Technical Degree': 6}
attrition['EducationField'] = attrition['EducationField'].map(d)

d = {'Female': 1, 'Male': 2}
attrition['Gender'] = attrition['Gender'].map(d)

d = {'Sales Executive': 8, 'Research Scientist': 7, 'Laboratory Technician': 3,
       'Manufacturing Director': 5, 'Healthcare Representative': 1, 'Manager': 4,
       'Sales Representative': 9, 'Research Director': 6, 'Human Resources': 2}
attrition['JobRole'] = attrition['JobRole'].map(d)

d = {'Single': 3, 'Married': 2, 'Divorced': 1}
attrition['MaritalStatus'] = attrition['MaritalStatus'].map(d)

d = {'Yes': 2, 'No': 1}
attrition['OverTime'] = attrition['OverTime'].map(d)

d = {'No': 0, 'Yes': 1}
attrition['Attrition'] = attrition['Attrition'].map(d)

attrition.dtypes
```

# **✅ Specifying the feature and target dataset**
***

```python
# define y
y = attrition['Attrition']
```

```python
# define X
X = attrition[['Age', 'BusinessTravel', 'DailyRate', 'Department',
       'DistanceFromHome', 'Education', 'EducationField','EmployeeNumber',
        'EnvironmentSatisfaction', 'Gender', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',
       'MaritalStatus', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
        'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears',
        'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']]
```

# **✅ Spilliting the training and testing dataset**
***

```python
# split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
```

```python
# verify shape
import numpy as np
a = np.array([X_train.shape, X_test.shape, y_train.shape, y_test.shape], dtype = 'object')
a
```

# **✅ Creating the model**
***

```python
# select model
from sklearn.ensemble import RandomForestClassifier 
model = RandomForestClassifier()
```

```python
# train model
model.fit(X_train, y_train)
```

# **✅ Prediction and evaluation of the model**
***

```python
# predict with model
y_pred = model.predict(X_test)
y_pred
```

```python
# model evaluation
from sklearn.metrics import mean_absolute_error
print("mean_absolute_error : ", mean_absolute_error(y_pred, y_test))
```

```python
# model accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```

```python
# model confusion matrix
from sklearn.metrics import confusion_matrix 
confusion_matrix(y_pred, y_test)
```

```python
# model classification report
from sklearn.metrics import classification_report 
print(classification_report(y_pred, y_test))
```
